````python
import dotenv
dotenv.load_dotenv()
from anthropic import Anthropic
import os
if os.getenv("ANTHROPIC_API_KEY") is None:
    raise Exception("ANTHROPIC_API_KEY environment variable not set")
client = Anthropic()
def add_user_message(messeges, content):
    messeges.append({"role": "user", "content": content})

def add_assistant_message(messeges, content):
    messeges.append({"role": "assistant", "content": content})

def chat_with_claude(messages,stop_sequences=None):
    if stop_sequences is not None:
        args={
            "model": "claude-sonnet-4-20250514",
            "max_tokens": 1024,
            "messages": messages,
            "stop_sequences": [stop_sequences]
        }
    else:
        args={
            "model": "claude-sonnet-4-20250514",
            "max_tokens": 1024,
            "messages": messages
    
        }
    response = client.messages.create(**args)
    return response.content[0].text
````

````python
messages = []

add_user_message(messages, "Generate a very short event bridge rule as json")
````

````python
response = chat_with_claude(messages)
print(response)
````

````python
response = chat_with_claude(messages)
print(response)
````

````python
messages = []

add_user_message(messages, "Generate a very short python sample function")
add_assistant_message(messages, "```python")
````

````python
response = chat_with_claude(messages,stop_sequences="```")
print(response)
````
