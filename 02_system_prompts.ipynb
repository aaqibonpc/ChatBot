```python
system_prompt = """
You are a patient math tutor.
Do not directly answer a student's questions.
Guide them to a solution step by step.
"""
```

```python
import dotenv
dotenv.load_dotenv()
from anthropic import Anthropic
import os
if os.getenv("ANTHROPIC_API_KEY") is None:
    raise Exception("ANTHROPIC_API_KEY environment variable not set")
client = Anthropic()
def add_user_message(messeges, content):
    messeges.append({"role": "user", "content": content})

def add_assistant_message(messeges, content):
    messeges.append({"role": "assistant", "content": content})

def chat_with_claude(messages):
    args={
        "model": "claude-sonnet-4-20250514",
        "max_tokens": 1024,
        "messages": messages,
        "system": system_prompt
    }
    response = client.messages.create(**args)
    return response.content[0].text
```

```python
messages=[]
while True:
    user_input = input()
    print("You:", user_input)
    print("-------------------------------")
    add_user_message(messages, user_input)
    answer = chat_with_claude(messages)
    add_assistant_message(messages, answer)
    print("Claude:",answer)
    print("-------------------------------")
```
